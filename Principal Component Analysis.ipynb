```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mne
```

# **PCA on Alcoholic Sample**

```python
path = '/content/drive/MyDrive/Depression Research/Alcoholic/alcohol1.csv'
df = pd.read_csv(path)
df = df[['FP2','F8','O2','FP1','F7','O1','F4','C4','P4','F3','C3','P3','FZ','CZ','PZ']]
#print(f)
#df = df.drop(['Unnamed: 0'],axis=1)
df_stats = df.describe()
df_stats = df_stats.transpose()

def norm(X):
  return X-df_stats['mean']/df_stats['std']

df = norm(df)
df = df.iloc[:250]
Xdash = df.values
index = np.where(np.isnan(Xdash).any(axis=1))[0]
Xdash = np.delete(Xdash,index,axis=0)
```

```python
Xdash.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
pca.fit(Xdash)
Xdash = pca.transform(Xdash)
```

```python
Xdash.shape
```
